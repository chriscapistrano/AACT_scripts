```python
import pandas as pd
```

```python
#read/open subject sheet
df = pd.read_csv('concat_df/2003.csv')
```

```python
df.head()
```

```python
#add columns from this sheet to df above (i.e., individual subjecst's sheet)
add_df = pd.read_csv('condONE.csv')
add_df.head()
```

```python
#let's add/merge
df['target'] = add_df['Target']
df['condition'] = add_df['ImageType']
```

```python
df.head()
```

```python
#remove the first 2 rows to have a cleaner dataset
df = df.drop(df.columns[[0, 1]], axis=1)

#rename the columns
df = df.rename(columns={'Unnamed: 1': 'Time', 'Unnamed: 2': 'CH14_value', 'Unnamed: 3': 'CH14_delta', 'Unnamed: 4': 'CH14_peak',
                        'Unnamed: 5': 'CH15_value', 'Unnamed: 6': 'CH15_delta', 'Unnamed: 7': 'CH15_peak'})


df.head()
```

```python
#save this as a new subject csv file
df.to_csv('concat_df/clean_data/2003.csv')
```
